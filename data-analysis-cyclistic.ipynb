```r
library(ggplot2)
library(tidyverse)
library(lubridate)
library(dplyr)
```

**Preparacion de los datos**

los datos de los viajes de cada mes, estan separados en archivos csv,lo primero sera combinar todos los datos en un solo dataframe que se llamará all_trip.

cargaremos los datos con los cuales se va a trabajar,los cuales corresponden a los viajes de los ultimos 12 meses de ambos tipos de clientes: ocasional y miembro.

Para ello utilizaremos la libreria Tidyverse la cual permite que se puedan leer los archivos csv y guardarlos en un dataframe.

Se crea un dataframe para cada mes del año desde enero hasta noviembre y se llamaran q1_2023 hasta q11_2023 

```r
q1_2023 <- read_csv("/kaggle/input/trip-data-cyclistic/202301-divvy-tripdata.csv")
q2_2023 <- read_csv("/kaggle/input/trip-data-cyclistic/202302-divvy-tripdata.csv")
q3_2023 <- read_csv("/kaggle/input/trip-data-cyclistic/202303-divvy-tripdata.csv")
q4_2023 <- read_csv("/kaggle/input/trip-data-cyclistic/202304-divvy-tripdata.csv")
q5_2023 <- read_csv("/kaggle/input/trip-data-cyclistic/202305-divvy-tripdata.csv")
q6_2023 <- read_csv("/kaggle/input/trip-data-cyclistic/202306-divvy-tripdata.csv")
q7_2023 <- read_csv("/kaggle/input/trip-data-cyclistic/202307-divvy-tripdata.csv")
q8_2023 <- read_csv("/kaggle/input/trip-data-cyclistic/202308-divvy-tripdata.csv")
q9_2023 <- read_csv("/kaggle/input/trip-data-cyclistic/202309-divvy-tripdata.csv")
q10_2023 <- read_csv("/kaggle/input/trip-data-cyclistic/202310-divvy-tripdata.csv")
q11_2023 <- read_csv("/kaggle/input/trip-data-cyclistic/202311-divvy-tripdata.csv")
```

**procesamiento y limpieza de datos**

Si se necesita analizar grandes cantidades de informacion, no es efectivo analizar cada archivo uno por uno, por eso creamos un único dataframe que contenga toda la informacion del año.

utilizaremos la funcion bind_rows la cual permite hacer coincidir el nombre de las columnas de todos los dataframes a procesar y unirlos en un unico dataframe que se llamará all_trip


```r
all_trip <- bind_rows(q1_2023,q2_2023, q3_2023, q4_2023,q5_2023,q6_2023,q7_2023,q8_2023,q9_2023,q10_2023,q11_2023)

head(all_trip)
```

Se eliminan las columnas start_lat,start_lng,end_lat,end_lng ya que no es necesaria esta informacion y se crea un nuevo dataframe organizado llamado all_trips

```r
all_trips <- all_trip %>%  
  select(-c(start_lat, start_lng, end_lat, end_lng))
head(all_trips)
```

Se procede a mostrar la informacion estadistica general del dataframe.

```r
colnames(all_trips) #verificamos los nombres de las columnas del df

summary(all_trips) #nos brinda informacion estadistica del df

nrow(all_trips) # numero de filas 

dim(all_trips)# numero de filas y columnas

str(all_trips)#verificacion de los tipos de datos del df
```

Para tener una mejor comprension de los datos hemos decidido cambiar el nombre de la variable en la columna member_casual, de esta forma "member" se cambiaria por "subscriber" y "casual" por "customer"

```r
#cambiamos el tipo de dato para que member sea suscriber y casual sea customer
all_trips <- all_trips %>% 
  mutate(member_casual = ifelse(member_casual=='member','subscriber','customer'))
head(all_trips)
```

Para verificar los viajes de los clientes y miembros de forma detallada es necesario separar los datos de fecha en varias columnas como mes, dia, año y dia de la semana en el dataframe, para ello se usan las funciones as.Date para convertir el dato de la columna started_at que estaba en string a fecha.
La funcion format, le da el formato de año, fecha y dia de la semana al dato de la fecha.

```r
#Se agregan nuevas columnas de fecha, mes, dia, año y dia de la semana.
all_trips$date <- as.Date(all_trips$started_at) #as.date nos permite convertir un string a formato de fecha
all_trips$month <- format(as.Date(all_trips$date), "%m") #la funcion format nos permite separar el dia, del mes y del año
all_trips$day <- format(as.Date(all_trips$date), "%d")
all_trips$year <- format(as.Date(all_trips$date), "%Y")
all_trips$day_of_week <- format(as.Date(all_trips$date), "%A")
head(all_trips)
```

Tambien se agrega una columna con el tiempo promedio de viaje de ambos tipos de usuarios.
Se utiliza la funcion difftime, la cual resta el tiempo de finalizacion del viaje con el tiempo inicial y brinda un dato de duracion en segundos.

```r
#restamos el tiempo de llegada menos el de inicio para calcular el tiempo promedio de viaje
all_trips$ride_length <- difftime(all_trips$ended_at,all_trips$started_at)
head(all_trips)
```

Ahora se valida si el tipo de dato de la columna ride_length es factor o numerico, es necesario que este numerico para poder realizar operaciones con los datos.

```r
#funciones para validar si el dato es factor o numerico
is.factor(all_trips$ride_length) #si es factor devuelve true y si no false
all_trips$ride_length <- as.numeric(as.character(all_trips$ride_length))# si es numerico devuelve true y sino false
is.numeric(all_trips$ride_length)
```

Se procede a filtrar los datos, empezando eliminando datos que sean negativos y que tengan el nombre de la estacion de inicio "HQ QR" para evitar que los datos esten cesgados o sean erroneos ya que esta estacion esta desactualizada.

 Ademas se eliminaran datos que esten faltantes o nulos y verificaremos que no hayan estos valores nulos con el comando is.na, que nos devuelve un TRUE si hay valores faltantes o FALSE si no los hay.
 
 
 Por ultimo  se obtendra un dataframe final organizado llamado df1

```r
#se eliminan datos negativos y con nombre de estacion hq qr
all_trips_v2 <- all_trips[!(all_trips$start_station_name == "HQ QR" | all_trips$ride_length<0),]
#se eliminan valores nulos y se almacenan en nuevo dataframe llamado df1
df1 <- na.omit(all_trips_v2)

is.na(df1) #devuelve TRUE si hay datos faltantes y FALSE si no los hay-
```

Luego se crea una nueva columna llamada ride_length_min para convertir el dato de ride_length que esta en segundos y pasarlo a minutos,asi será mas compresible al momento de visualizar los datos.

```r
df1 <- df1 %>%
  mutate(ride_length_min=num(ride_length/60))
head(df1)
```

**proceso de analisis de datos**

En este proceso de analisis se realiza una comparacion entre los datos con la funcion agregate, el cual nos muestra  cuando tarda cada tipo de cliente en sus viajes en promedio, el valor maximo y minimo y la media.

```r
#comparo el tiempo que tarda uin suscriptor que el de un cliente casual
aggregate(df1$ride_length ~ df1$member_casual, FUN = mean)
aggregate(df1$ride_length ~ df1$member_casual, FUN = median)
aggregate(df1$ride_length ~ df1$member_casual, FUN = max)
aggregate(df1$ride_length ~ df1$member_casual, FUN = min)
```

Se puede evidenciar en los datos que el tipo de usuario "Cliente" tiene una duracion de viaje promedio mas alta que el tipo de usuario "Subscriber"

Ahora se evidencia como se comporta a lo largo del tiempo verificando el tiempo promedio que tarda cada cliente  por dia de la semana.

```r
#Ver el tiempo promedio de viaje cada día para miembros y usuarios ocasionales                   
aggregate(df1$ride_length ~ df1$member_casual + df1$day_of_week, FUN = mean)

#se organizan los dias de la semana
df1$day_of_week <- ordered(df1$day_of_week, levels=c("Sunday", "Monday", "Tuesday", "Wednesday", "Thursday", "Friday", "Saturday"))

#se compara de nuevo el tiempo promedio por dia de la semana para cliente y suscriptor
aggregate(df1$ride_length ~ df1$member_casual + df1$day_of_week, FUN = mean )
```

Con estas comparaciones que se han realizado con la funcion agregate se puede notar que el tipo de usuario subscriptor tiene un tiempo promedio de viaje mas bajo que el cliente ocasional.

Luego se organizan los datos para mostrar la informacion de los tipos de usuario por dia de la semana.

```r
#analizar datos de usuarios por tipo y día de la semana
df1 %>% 
  mutate(weekday = wday(started_at, label = TRUE)) %>%  #creo un objeto wday para cada dia de la semana()
  group_by(member_casual, weekday) %>%  #agrupo tipo de usuario y dia de la semana
  summarise(number_of_rides = n(),average_duration = mean(ride_length)) %>% #se crean columnas de numero de viajes y promedio de viaje
  arrange(member_casual, weekday)	#se ordenan los datos de tipo de usuario y dia de la semana
```

**Visualizacion de los datos**

Ahora se procede a visualizar los datos para ver la relacion entre cada tipo de usuario y el numero de viajes que ha realizado por cada dia de la semana.

```r
#Visualicemos el número de viajes por tipo de ciclista.
df1 %>% 
  mutate(weekday = wday(started_at, label = TRUE)) %>% 
  group_by(member_casual, weekday) %>% 
  summarise(number_of_rides = n(),average_duration = mean(ride_length)) %>% 
  arrange(member_casual, weekday)  %>% 
  ggplot(aes(x = weekday, y = number_of_rides, fill = member_casual,title='number of rides vs weekday')) + #grafica de dia de la semana vs numero de viajes categorizado por tipo de cliente
  geom_col(position = "dodge")
```

En la siguiente visualizacion se compara el tiempo promedio de cada tipo de usuario por semana.

```r
# ahora compararemos dia de la semana con duracion de viaje

df1 %>%
  mutate(weekday = wday(started_at, label = TRUE)) %>% 
  group_by(member_casual, weekday) %>% 
  summarise(number_of_rides = n(),average_duration = mean(ride_length)) %>% 
  arrange(member_casual, weekday)  %>% 
  ggplot(aes(x = weekday, y = average_duration, fill = member_casual)) + #grafica de dia de la semana vs numero de viajes categorizado por tipo de cliente
  geom_col(position = "dodge")+ labs(title = "average_duration vs weekday")
```

Ademas se visualizara la preferencia del tipo de bicicleta para cada usuario.

```r
preference_bike <- df1 %>%
  group_by(member_casual, rideable_type) %>%
  summarize(sum_rideable_type = n())

head(preference_bike, 3)

ggplot(preference_bike, aes(x = member_casual, y = sum_rideable_type, fill = rideable_type)) +geom_bar(stat = "sum",position = "dodge") +labs(x = "member_casual", y = "rideable_type", fill = "member_casual", title = "rideable_type vs member_casual")
```

```r
Por ultimo se procede a guardar un resumen de los datos en un nuevo dataframe que se llamará resumen_bicicletas y se cargará a un archivo csv llamado resumen_biciletas.csv
```

```r
resumen_biciletas <- aggregate(df1$ride_length ~ df1$member_casual + df1$day_of_week, FUN = mean)

write.csv(resumen_biciletas, file='/kaggle/working/resumen_biciletas.csv')
```
